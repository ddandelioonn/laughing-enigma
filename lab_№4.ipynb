```python
#класс Human

class Human: #1
    default_name = "NoName" #2
    default_age = 0
    
    def __init__(self, name=default_name, age=default_age): #3
        self.name = name
        self.age = age
        self._money = 0
        self._house = None
        
    def info(self): #4
        print('Name:', self.name)
        print('Age:', self.age)
        print('Money:', self._money)
        print('House:', self._house)
    
    @staticmethod #5
    def default_info(): 
        print('Default Name:', Human.default_name)
        print('Default Age:', Human.default_age)
    
    def _make_deal(self, house, discount): #6
        self._money -=  house.final_price(discount)
        self._house = house
        
    def earn_money(self, money): #7
        self._money += money
    
    def buy_house(self, house, discount): #8
        if self._money >= house.final_price(discount):
            self._make_deal(house, discount)
            print("Поздравляем с покупкой дома!")
        else:
            print("Недостаточно денег для покупки дома")
    
    
#класс House

class House: #9
    def __init__(self, area, price):
        self.area = area
        self.__price = price
    
    def final_price(self, discount):
        return self.__price * (100 - discount) / 100

#класс SmallHouse

class SmallHouse(House): #10
    def __init__(self, price):
        super().__init__(40, price)

#тесты

#1 вызываем метод default_info() для класса Human
Human.default_info()

#2 создаем объект класса Human
human = Human('Bob', 22)

#3 выводим информацию о созданном объекте
human.info()

#4 создаем объект класса SmallHouse
small_house = SmallHouse(10000)

#5 пытаемся купить дом 
human.buy_house(small_house, 10)

#6 увеличиваем количество денег
human.earn_money(100000)

#7 пытаемся купить дом еще раз
human.buy_house(small_house, 10)

#8 смотрим, что поменялось в жизни Боба после покупки дома
human.info() 
```
